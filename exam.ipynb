```python
def solution(numbers):
    """
    Check all consecutive triples in the array for zigzag pattern.
    
    A triple (a, b, c) is a zigzag if either:
    - a < b > c (peak)
    - a > b < c (valley)
    
    Args:
        numbers: List of integers
        
    Returns:
        List of 1s and 0s indicating if each triple is a zigzag
    """
    # If array has less than 3 elements, return empty array
    if len(numbers) < 3:
        return []
    
    result = []
    
    # Check each consecutive triple
    for i in range(len(numbers) - 2):
        a = numbers[i]
        b = numbers[i + 1]
        c = numbers[i + 2]
        
        # Check if it's a zigzag: either a < b > c OR a > b < c
        is_zigzag = (a < b > c) or (a > b < c)
        
        result.append(1 if is_zigzag else 0)
    
    return result


# Test cases from the problem
if __name__ == "__main__":
    # Test case 1: [1, 2, 1, 3, 4]
    test1 = [1, 2, 1, 3, 4]
    print(f"Input: {test1}")
    print(f"Output: {solution(test1)}")
    print(f"Expected: [1, 1, 0]")
    print()
    
    # Test case 2: [1, 2, 3, 4]
    test2 = [1, 2, 3, 4]
    print(f"Input: {test2}")
    print(f"Output: {solution(test2)}")
    print(f"Expected: [0, 0]")
    print()
    
    # Test case 3: [1000000000, 1000000000, 1000000000]
    test3 = [1000000000, 1000000000, 1000000000]
    print(f"Input: {test3}")
    print(f"Output: {solution(test3)}")
    print(f"Expected: [0]")
    print()
    
    # Additional test cases
    test4 = [3, 1, 2]
    print(f"Input: {test4}")
    print(f"Output: {solution(test4)}")
    print(f"Expected: [1] (3 > 1 < 2)")
```
