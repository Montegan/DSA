```python
def findsum(arr:list[int])->tuple[int,int,int]:
    sum=0
    max=arr[0]
    min=arr[0]

    for i in arr:
        sum=sum+i
        if i>max:
            max=i
        if i<=min:
            min=i
    return (sum,max,min)
```

```python
# leetcode 169 

class Solution:
    def majorityElement(self, nums: List[int]) -> int:
        freq=dict()
        max=0
        maxkey=0

        for n in nums:
            freq[n]=freq.get(n,0)+1
        for i in freq.keys():
            if freq[i] > max:
                max=freq[i]
                maxkey=i
        return maxkey
```

```python
# n^2 solution
class Solution:
    def smallerNumbersThanCurrent(self, nums: List[int]) -> List[int]:
        final_count=[0]*len(nums)
        i=0
        while i<len(nums):
            j=i+1
            while j<len(nums):
                if nums[i]>nums[j]:
                    final_count[i]+=1
                elif nums[i]<nums[j]:
                    final_count[j]+=1
                j+=1
            i+=1
        return final_count
```

```python
# leetcode maximum subarray sum
# Kadence algorithm solution.

class Solution:
    def maxSubArray(self, nums: List[int]) -> int:
        current_sum=nums[0]
        max_sum=nums[0]
        for i in range(1,len(nums)):
            current_sum= max(current_sum + nums[i],nums[i])
            max_sum=max(current_sum,max_sum)
        return max_sum
```

```python
# naive solution to max subarray sum

class Solution:
    def maxSubArray(self, nums: List[int]) -> int:
        curr_sum=0
        max_sum=nums[0]
        for i in nums:
            curr_sum+=i
            if curr_sum<i:
                curr_sum=i
            max_sum=max(max_sum,curr_sum)
        return max_sum
```

```python
class Solution:
    def maxSumDivThree(self, nums: List[int]) -> int:
        prefix_sum=0
        previous_sums={}


        for i in nums:
            current_div=i%3
            if current_div == 0:
                prefix_sum+=i
            elif 3-current_div in previous_sums:
                if previous_sums.get(current_div,0)<i:
                    prefix_sum=prefix_sum-previous_sums.get(current_div,0) + previous_sums.get(3-current_div,0)
                    prefix_sum=prefix_sum+i+ previous_sums.get(3-current_div,0)
                    previous_sums[current_div]=i
                if previous_sums.get(current_div,0)>i:
                    prefix_sum=prefix_sum
            else:
                previous_sums[current_div]=i
        return prefix_sum
```

Leetcode distribute candies problem 

```python
class Solution:
    def distributeCandies(self, candyType: List[int]) -> int:
        set_candy=set(candyType)
        n=(len(candyType)//2)
        count=0
        i=0
        j=0
        if len(set_candy) <= n:
            return len(set_candy)
        
        while i<n:
            if candyType[j] in set_candy:
                count+=1
                set_candy.remove(candyType[j])
                i+=1
            j+=1
        return count
```
